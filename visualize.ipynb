```python
#area plot
import matplotlib.pyplot as plt
import pandas as pd

data={
    'Apples':[1,2,3,4,5,6],
    'Banana':[4,3,1,2,3,4],
    'Grapes':[1,2,3,4,5,6]
}

df=pd.DataFrame(data,index=['Mon','Tue','Wed','Thu','Fri','Sat'])
df.plot.area()

plt.title("Graphs")
plt.xlabel("Day")
plt.ylabel("Sold items")
plt.show()
```

```python
#bar graph
import pandas as pd
import matplotlib.pyplot as plt

data={
     'Apples':  [3, 4, 2, 5, 6],
    'Bananas': [2, 3, 4, 2, 1],
    'Cherries':[1, 2, 1, 3, 2]
}
df = pd.DataFrame(data, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'])

df.plot(kind='bar')
plt.title('Bar graph')
plt.xlabel('Day')
plt.ylabel('Quantity sold')
plt.show()
```
