```python
import pandas as pd

csv=pd.read_csv("data_file.csv")
print("Dataset")
print(csv.head())
age=csv[['Name','Age']]
print(age)
#selecting a singal row
csv=csv.set_index(['Name'])
row=csv.loc['Krishiv']

#selecting multiple rows
row=csv.loc[['Krishiv','Snehal']]
print(row)

#selecting rows and cols
selection=csv.loc[['Krishiv','Snehal'],['City','Score']]
print(selection)

#selecting all rows and specific colums
csv=csv.reset_index()
specific=csv.loc[:,['Name','City','Score']]
print(specific)

#The .iloc[] function is used for position-based indexing.
#It allows us to access rows and columns by their integer positions. 
#It is similar to .loc[] but only accepts integer-based indices to specify rows and columns.
row=csv.iloc[3]
print(row)

#selecting multiple columns with the indexes
row=csv.iloc[[4,5,3]]
print(row)

#selecting specific rows and columns by position
row=csv.iloc[[3,5],[1,2]]
print(row)

#selecting all rows and specific colums
row=csv.iloc[:,[1,2]]
print(row)

#head
print(csv.head(5))

#tail
print(csv.tail(5))

#get specific value from the name 
csv=csv.set_index('Name')
val=csv.at['Snehal','Age']
print(val)

#query()
query=csv.query("Age>25 and City=='Surat'")
print(query)
```

```python
#slicing in pandas

import pandas as pd
player_list = [['M.S.Dhoni', 36, 75, 5428000],
               ['A.B.D Villers', 38, 74, 3428000],
               ['V.Kohli', 31, 70, 8428000],
               ['S.Smith', 34, 80, 4428000],
               ['C.Gayle', 40, 100, 4528000],
               ['J.Root', 33, 72, 7028000],
               ['K.Peterson', 42, 85, 2528000]]

df=pd.DataFrame(player_list,columns=['Name','Age','Weight','Salary'])
print(df)

#slicing using iloc
df1=df.iloc[:3]
print(df1)

#slicing colums
df1=df.iloc[:,0:2]
print(df1)

#slicing specific cell
val=df.iloc[2,3]
print(val)

#slicing based on condition
cond=df[df['Age']>40].iloc[:,:]
print(cond)
```

```python
#iterating over rows and colums

import pandas as pd
import numpy as np

dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}
df=pd.DataFrame(dict)
print(df)

for i,j in df.iterrows():
    print(i,j)
    print()

cols=list(df)

for i in cols:
    print(df[i][2:5])

print(df.sample)
```

```python
#grouping in pandas
import numpy as np
import pandas as pd

data = {
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Salary': [50000, 60000, 55000, 58000, 62000]
}

df=pd.DataFrame(data)
print(df)

#group by one column
df.groupby('Department')['Salary'].mean()

#multiple aggregation
df.groupby('Department')['Salary'].agg(['min','max','mean','sum'])

#groupby multiple columns
df.groupby(['Department','Name'])['Salary'].sum()

#count
df.groupby('Department').count()
```
